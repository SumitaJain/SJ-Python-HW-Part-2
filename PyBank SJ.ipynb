```python
# Building off of downloaded hints provided by instructors/TAs (copied whole hint at bottom)

# Dependencies
import csv
import os

# Files to load and output (Remember to change these)
file_to_load = os.path.join("Resources", "budget_data.csv")
file_to_output = os.path.join("budget_analysis.txt")
```

```python
# Track various financial parameters
total_months = 0
month_of_change = []
net_change_list = []
greatest_increase = ["", 0]
greatest_decrease = ["", 9999999999999999999]
total_netPL = 0
```

```python
# Read the csv and convert it into a list of dictionaries
with open(file_to_load) as financial_data:
    reader = csv.reader(financial_data)

    # Read the header row
    header = next(reader)

    # Extract first row to avoid appending to net_change_list
    # YOUR CODE HERE
    row_1 = next(reader)
    prev_netPL = int(row_1[1])
    
    for row in reader:

        # Track the total
        # YOUR CODE HERE
        total_months = total_months + 1
        total_netPL = total_netPL + int(row_1[1])
        
        # Track the net change
        # YOUR CODE HERE
        netPL_change = int(row[1]) - prev_netPL
        prev_netPL = int(row_1[1])
        net_change_list.append(netPL_change)
        month_of_change = month_of_change + [row[0]]

        # Calculate the greatest increase
        # YOUR CODE HERE
        if netPL_change > greatest_increase[1]:
            greatest_increase[0] = row[0]
            greatest_increase[1] = netPL_change

        # Calculate the greatest decrease
        # YOUR CODE HERE
        if netPL_change < greatest_decrease[1]:
            greatest_decrease[0] = row[0]
            greatest_decrease[1] = netPL_change
```

```python
# Calculate the Average Net Change
# YOUR CODE HERE
netPL_average_monthly_change = sum(net_change_list)/len(net_change_list)
```

```python
# Generate Output Summary
output = (
    f"\nFinancial Analysis\n"
    f"----------------------------\n"
    f"Total Months: {total_months}\n"
    f"Total: ${total_netPL}\n"
    f"Average  Change: ${netPL_average_monthly_change:.2f}\n"
    f"Greatest Increase in Profits: {greatest_increase[0]} (${greatest_increase[1]})\n"
    f"Greatest Decrease in Profits: {greatest_decrease[0]} (${greatest_decrease[1]})\n")

```

```python
# Print the output (to terminal)
# YOUR CODE HERE
print(output)

# Export the results to text file
# YOUR CODE HERE
with open(file_to_output, "w") as txt_file:
    txt_file.write(output)
```
